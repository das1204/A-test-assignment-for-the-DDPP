```python
import functions as fn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
import xgboost as xgb
```

```python
df = fn.get_CPI_data_from_rosstat(type_CPI='01').merge(
    fn.get_CPI_data_from_rosstat(type_CPI='02')[['', fn.CPI_guide['02'][0]]],
    on='',
    how='left').merge(fn.get_CPI_data_from_rosstat(type_CPI='03')[['', fn.CPI_guide['03'][0]]],
                      on='',
                      how='left').merge(fn.get_CPI_data_from_rosstat(type_CPI='04')[['', fn.CPI_guide['04'][0]]],
                                        on='',
                                        how='left')
df[''] = pd.to_datetime(df[''])

df2 = fn.get_DollarER_data_from_cbr()

df3 = fn.get_KR_data_from_cbr()

df_glob = df.merge(df2, on='', how='left').merge(df3, on='', how='left')

df_glob = df_glob.dropna()
df_glob[''] = pd.DatetimeIndex(df_glob['']).year
df_glob[''] = pd.DatetimeIndex(df_glob['']).month

df_fc = df_glob.copy()[['', '  ', ' ', ' _MoM', '', '', '']]
df_fc = df_fc.set_index('')

corr = df_fc.corr()['  ']
pd.DataFrame(corr)
corr = df_fc.corr()
plt.figure(figsize=(5, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', cbar=False)
plt.tight_layout(pad=1.0)
```

```python
df_val = df_glob[-6:]
df_tt = df_glob[:-6]

X_val, y_val = df_val[-6:].loc[:, [' _MoM', '']].values, df_val[-6:].loc[:, ['    ']].values

X, y = df_tt.loc[:, [' _MoM', '']].values, df_tt.loc[:, ['    ']].values

X_train = X[:int(len(X) * 0.8)]
y_train = y[:int(len(X) * 0.8)]

X_test = X[int(len(X) * 0.8):]
y_test = y[int(len(X) * 0.8):]
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

lr_predictions = lr.predict(X_test)

print("MSE: %f" % (mean_squared_error(y_test, lr_predictions)))

print("R2 Score: %f" % (r2_score(y_test, lr_predictions)))

print("MAPE: %f" % (np.mean(np.abs((y_test - lr_predictions) / y_test)) * 100))
```

```python
import pickle

# model -   
with open('04.pkl', 'wb') as f:
    pickle.dump(lr, f)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

rf_predictions = rf.predict(X_test)

print("MSE: %f" % (mean_squared_error(y_test, rf_predictions)))

print("R2 Score: %f" % (r2_score(y_test, rf_predictions)))

print("MAPE: %f" % (np.mean(np.abs((y_test - rf_predictions) / y_test)) * 100))
```

```python
XGB = xgb.XGBRegressor(objective='reg:squarederror')
XGB.fit(X_train, y_train)

XGB_predictions = XGB.predict(X_test)

print("MSE: %f" % (mean_squared_error(y_test, XGB_predictions)))

print("R2 Score: %f" % (r2_score(y_test, XGB_predictions)))

print("MAPE: %f" % (np.mean(np.abs((y_test - XGB_predictions) / y_test)) * 100))
```

```python
from sklearn.model_selection import GridSearchCV

#  
model = xgb.XGBRegressor()

#   ,    
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_child_weight': [1, 3, 5],
    'gamma': [0.0, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

#   GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', n_jobs=-1)

#  GridSearchCV   
grid_search.fit(X_train, y_train)

#      
print(" :", grid_search.best_params_)
```

```python
XGB_p = xgb.XGBRegressor(objective='reg:squarederror',
                         colsample_bytree=1.0,
                         gamma=0.0,
                         learning_rate=0.01,
                         max_depth=3,
                         min_child_weight=5,
                         n_estimators=100,
                         subsample=0.9)

XGB_p.fit(X_train, y_train)

XGB_p_predictions = XGB_p.predict(X_test)

print("MSE: %f" % (mean_squared_error(y_test, XGB_p_predictions)))

print("R2 Score: %f" % (r2_score(y_test, XGB_p_predictions)))

print("MAPE: %f" % (np.mean(np.abs((y_test - XGB_p_predictions) / y_test)) * 100))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(df_tt[''][-30:], df_tt['    '][-30:], label="Fact")
plt.plot(df_tt[''][-len(lr_predictions):], lr_predictions, label="lr")
plt.plot(df_tt[''][-len(rf_predictions):], rf_predictions, label="rf")
plt.plot(df_tt[''][-len(XGB_predictions):], XGB_predictions, label="XGB")
plt.plot(df_tt[''][-len(XGB_p_predictions):], XGB_p_predictions, label="XGB_p")
plt.legend(loc="best")
plt.grid(True)
```
